# Popularity Based Recommendation System

```python
# merging 2 files
merged_df = pd.merge(goodreads_work, goodreads_reviews, on='work_id', how='inner')
merged_df.info()
```

```python
#convert columns to string before filling strig values
merged_df["isbn13"] = merged_df["isbn13"].astype(str)
merged_df["isbn"] = merged_df["isbn"].astype(str)
```

```python
# data frame wth respect to rating 
merged_df.groupby('original_title').count()[['rating']]
```

```python
# building the df for popularity
popularity_df = merged_df.groupby('original_title').agg({
    'author': 'first',
    'original_publication_year': 'first',
    'genres': 'first',
    'image_url': 'first',
    'ratings_count': 'count',
    'avg_rating': 'mean'
}).reset_index()

# only taking books with reasonable rating count
popularity_df = popularity_df[popularity_df['ratings_count'] >= 100]
popularity_df = popularity_df.sort_values('avg_rating', ascending=False).reset_index(drop=True)
popularity_df
```

```python
# checking the code working
popularity_df = popularity_df[popularity_df["genres"].str.contains("fantasy", case=False, na=False)]
popularity_df

```

```python
popularity_df['image_url'][0]
```

# Collaborative Filtering Based Recommendation System

```python
user_rating_counts = merged_df.groupby("user_id")["rating"].count()
active_users = user_rating_counts[user_rating_counts >= 100]
active_users = active_users[active_users].index
```

```python
filtered_rating = merged_df[merged_df["user_id"].isin(active_users)].reset_index(drop=True)
filtered_rating
```

```python
# filtering the data frame on the basis pf their book title
y = filtered_rating.groupby('original_title')["rating"].count() >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['original_title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='original_title', columns ='user_id', values='rating')
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
# building function for book recommendation
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = merged_df[merged_df['original_title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('original_title')['original_title'].values))
        item.extend(list(temp_df.drop_duplicates('original_title')['author'].values))
        item.extend(list(temp_df.drop_duplicates('original_title')['image_url'].values))

        data.append(item)
    
    #print(data)
    return data
    #return suggestions




```

```python
recommend('Winter')
```
